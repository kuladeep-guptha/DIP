```python
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def psnr(img1, img2):
    mse = np.mean((img1 - img2) ** 2)
    if mse == 0:
        return 100
    PIXEL_MAX = 255.0
    return 10 * math.log10((PIXEL_MAX)**2 / mse)
```

```python
img1 = cv2.imread('kula.jpg')
img2 = cv2.imread('kula.jpg')
img2 = cv2.blur(img2, (10,10))
```

```python
print(psnr(img1, img2))
```

```python
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
cv2.imwrite('blurred-face.jpeg', img2)
```

```python
mean=0
standard_deviation=180
noise = np.zeros(img2.shape, np.uint8)
cv2.randn(noise, mean, standard_deviation)
img3 = cv2.add(img2, noise)
```

```python
plt.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))
plt.title("Noisy Image")
plt.show()
```

```python
print(psnr(img1,img3))
```

```python
img1 = cv2.imread('kula.jpg')
cv2.imwrite('kula.jp2', img1, [int(cv2.IMWRITE_JPEG2000_COMPRESSION_X1000), 70])
img2 = cv2.imread('kula.jp2')
print(psnr(img1, img2))
```
